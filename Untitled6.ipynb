```python
"""
Global Solution - Modelagem Linear para Aprendizagem de Máquina
Análise estatística de missões espaciais - base Kaggle
"""

import os
import pandas as pd
import matplotlib.pyplot as plt

ARQUIVO = "space_missions_kaggle_amostra.csv"
PASTA_SAIDA = "resultados"
os.makedirs(PASTA_SAIDA, exist_ok=True)

df["data_lancamento"] = pd.to_datetime(df["Datum and Time of Launch"], errors="coerce", utc=True)
df = df.dropna(subset=["data_lancamento"]).copy()
df["ano_lancamento"] = df["data_lancamento"].dt.year
df["mes_lancamento"] = df["data_lancamento"].dt.month
primeira_data = df["data_lancamento"].min()
df["dias_desde_inicio"] = (df["data_lancamento"] - primeira_data).dt.total_seconds() / 86400

def estatistica_descritiva(serie):
    return pd.Series({
        "media": serie.mean(),
        "mediana": serie.median(),
        "moda": serie.mode().iloc[0] if not serie.mode().empty else None,
        "maximo": serie.max(),
        "minimo": serie.min(),
        "amplitude": serie.max() - serie.min(),
        "variancia": serie.var(),
        "desvio_padrao": serie.std(),
        "q1": serie.quantile(0.25),
        "q2": serie.quantile(0.50),
        "q3": serie.quantile(0.75),
    })

print("BASE DE DADOS - PRIMEIRAS LINHAS")
print(df.head())

freq_ano = df["ano_lancamento"].value_counts().sort_index().reset_index()
freq_ano.columns = ["ano_lancamento", "frequencia"]
freq_ano["frequencia_relativa_%"] = (freq_ano["frequencia"] / len(df) * 100).round(2)
freq_ano.to_csv(os.path.join(PASTA_SAIDA, "frequencia_ano_lancamento.csv"), index=False)
print("\nTABELA DE FREQUENCIA - ANO DE LANCAMENTO")
print(freq_ano)

classes_dias = pd.cut(df["dias_desde_inicio"], bins=5)
freq_dias = classes_dias.value_counts().sort_index().reset_index()
freq_dias.columns = ["intervalo_dias_desde_inicio", "frequencia"]
freq_dias["frequencia_relativa_%"] = (freq_dias["frequencia"] / len(df) * 100).round(2)
freq_dias.to_csv(os.path.join(PASTA_SAIDA, "frequencia_dias_desde_inicio.csv"), index=False)
print("\nTABELA DE FREQUENCIA - DIAS DESDE O INICIO")
print(freq_dias)

plt.figure(figsize=(8, 5))
freq_status = df["Status of the mission"].value_counts()
plt.bar(freq_status.index, freq_status.values)
plt.title("Distribuicao do Status das Missoes")
plt.xlabel("Status da missao")
plt.ylabel("Quantidade de lancamentos")
plt.tight_layout()
plt.savefig(os.path.join(PASTA_SAIDA, "grafico_status_missoes.png"), dpi=150)
plt.close()

plt.figure(figsize=(8, 5))
plt.hist(df["dias_desde_inicio"], bins=5, edgecolor="black")
plt.title("Distribuicao Temporal dos Lancamentos")
plt.xlabel("Dias desde o primeiro lancamento da amostra")
plt.ylabel("Frequencia")
plt.tight_layout()
plt.savefig(os.path.join(PASTA_SAIDA, "histograma_dias_desde_inicio.png"), dpi=150)
plt.close()

estat_ano = estatistica_descritiva(df["ano_lancamento"])
estat_dias = estatistica_descritiva(df["dias_desde_inicio"])
estatisticas = pd.DataFrame({
    "ano_lancamento": estat_ano,
    "dias_desde_inicio": estat_dias,
})
estatisticas.to_csv(os.path.join(PASTA_SAIDA, "estatisticas_descritivas.csv"))
print("\nESTATISTICAS DESCRITIVAS")
print(estatisticas)

resumo_status = df["Status of the mission"].value_counts(normalize=True).mul(100).round(2)
resumo_empresas = df["Company Name"].value_counts().head(10)
print("\nPercentual por status de missao (%)")
print(resumo_status)
print("\nTop 10 empresas por quantidade de lancamentos")
print(resumo_empresas)
```
